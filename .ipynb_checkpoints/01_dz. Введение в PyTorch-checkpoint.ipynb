# Тема: Введение в PyTorch

```python
import numpy as np

import torch
import torchvision
from torch import nn 
from torch import optim
import torchvision.transforms as transforms
import matplotlib.pyplot as plt

import torch.nn.functional as F
```

### 1. Добиться вразумительного результата от модели, собранной на занятии (7 угаданныx картинок из 10 предложенных на семинаре). 

Варианты изменений:
* изменение слоёв и их количества;
* изменение метода оптимизации; 
* изменение процесса обучения; 
* \*преобразование данных transform


```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

```python
train_dataset = torchvision.datasets.CIFAR10(
    root='../data/',  # путь, куда будут скачиваться данные(если download=true), 
                   # или где лежат данные(dowload=False)
    train=True,    # говорим что используем датасет для обучения  
    transform=transforms.ToTensor(), # преобразуем данные в формат тензоров
    download=True
)


image, label = train_dataset[0]
print(image.shape)
print(label)
```

```python
classes = ['plane', 'car', 'bird', 'cat','deer', 'dog', 
           'frog', 'horse', 'ship', 'truck']
```

```python
train_loader = torch.utils.data.DataLoader(
    dataset=train_dataset,
    batch_size=64, 
    shuffle=True
)
```

```python
for imgs, labels in train_loader:
    print(imgs.shape, labels.shape)
    print(classes[labels[0]])
    plt.imshow(imgs[0].permute(1, 2, 0))
    break
```

```python
test_dataset = torchvision.datasets.CIFAR10(
    root='../data/',
    train=False,
    download=True,
    transform=transforms.ToTensor()
)

test_loader = torch.utils.data.DataLoader(
    test_dataset,
    batch_size=8,
    shuffle=False
)
```

```python
class Net(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels=3,
                               out_channels=6,
                               kernel_size=3)
        self.pool = nn.MaxPool2d(kernel_size=2,
                                 stride=2)
        self.conv2 = nn.Conv2d(in_channels=6,
                               out_channels=16,
                               kernel_size=3)
        # изменение слоёв и их количества
        self.conv3 = nn.Conv2d(in_channels=16,
                               out_channels=32,
                               kernel_size=3)
        self.fc1 = nn.Linear(in_features=32 * 2 * 2,
                             out_features=90)
        
        self.fc2 = nn.Linear(in_features=90,
                             out_features=60)
        self.fc3 = nn.Linear(in_features=60,
                             out_features=30)
        self.fc4 = nn.Linear(in_features=30,
                             out_features=10)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = self.pool(F.relu(self.conv3(x)))
        x = x.reshape(-1, 32 * 2 * 2)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.relu(self.fc3(x))
        x = self.fc4(x)
        return x


net = Net().to('cuda')
```

```python
net
```

```python
criterion = nn.CrossEntropyLoss()
# изменение метода оптимизации
optimizer = optim.Adam(net.parameters(), lr=0.001)
```

```python
%%time
num_epochs = 20

for epoch in range(num_epochs):
    running_loss = 0.0  # Для вывода процесса обучения
    running_items = 0.0

    for i, data in enumerate(train_loader):
        inputs, labels = data[0].to(device), data[1].to(device)
        # Обнуляем градиент
        optimizer.zero_grad()
        # Делаем предсказание
        outputs = net(inputs)
        # Рассчитываем лосс-функцию
        loss = criterion(outputs, labels)
        # Делаем шаг назад по лоссу
        loss.backward()
        # Делаем шаг нашего оптимайзера
        optimizer.step()

        # выводим статистику о процессе обучения
        running_loss += loss.item()
        running_items += len(labels)
        if i % 300 == 0:    # печатаем каждые 300
            print(f'Epoch [{epoch + 1}/{num_epochs}]. ' \
                  f'Step [{i + 1}/{len(train_loader)}]. ' \
                  f'Loss: {running_loss / running_items:.3f}')
            running_loss, running_items = 0.0, 0.0

print('Training is finished!')
```

```python
net.to('cpu')
```

```python
data_iter = iter(test_loader)
images, labels = data_iter.next()
```

```python
outputs = net(images)
imgs = torchvision.utils.make_grid(images)
plt.figure(figsize=(10, 5))
plt.imshow(imgs.permute(1, 2, 0).numpy())
print('GroundTruth: ', ' '.join(classes[labels[j]] for j in range(len(labels))))
```

```python
_, predicted = torch.max(outputs, 1)

predicted
```

```python
gt = np.array([classes[labels[j]] for j in range(len(labels))])
pred = np.array([classes[predicted[j]] for j in range(len(labels))])

print(gt)
print(pred)
print(f'Accuracy is {(gt == pred).sum() / len(gt)}')
```
